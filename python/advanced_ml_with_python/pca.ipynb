```python
import numpy as np
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.cm as cm
```

```python
digits = load_digits()
data = scale(digits.data)

n_samples, n_features = data.shape
n_digits = len(np.unique(digits.target))
labels = digits.target
```

```python
pca = PCA(n_components=10)
data_r = pca.fit(data).transform(data)
lda = LDA(n_components=2)
data_r2 = lda.fit(data, labels).transform(data)

# Print the variance explained by each of the first ten principal components ordered by explanatory power.
print('Explained variance ratio (first two components): %s' %str(pca.explained_variance_ratio_))
print('Sum of explained variance (first two components): %s' %str(sum(pca.explained_variance_ratio_)))
```
