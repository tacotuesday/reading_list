## Predicting regression

### Linear Regression

```python
# `df_bikes['cnt']` is the target variable. It is the number of bikes rented in a given day.
# `cnt` is the sum of `casual` and `registered`. We can drop these two columns.
df_bikes = df_bikes.drop(['casual', 'registered'], axis=1)
```

```python
# Save the data for future use.
df_bikes.to_csv('bike_rentals_cleaned.csv', index=False)
```

```python
# Declaring predictor and target columns. `cnt`, the target variable, is the last column.
X = df_bikes.iloc[:, :-1]
y = df_bikes.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
# Ignore warnings in this specific notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# Initialize the model
lin_reg = LinearRegression()

# Fit the model
lin_reg.fit(X_train, y_train)

# Make predictions for the test set
y_pred = lin_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'RMSE: {rmse}')
```

```python
# We can't understand the RMSE without understanding the spread of `cnt` values.
df_bikes['cnt'].describe()

# The RMSE is 1/3 of the mean value of `cnt`. This is a good result.
# Compared to the scale of the data, the error is relatively small.
# An RMSE smaller than the standard deviation indicates that the predictions are, on average, closer to the actual values than a naive prediction.
```

### XGBoost

```python
# Let's compare this to XGBoost
# Import XGBRegressor
from xgboost import XGBRegressor

# Initialize the model
xgb_reg = XGBRegressor()

# Fit the model on the training set
xgb_reg.fit(X_train, y_train)

# Make predictions for the test set
y_pred = xgb_reg.predict(X_test)
```

```python
# Calculate the RMSE
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(f'RMSE: {rmse}')
```

### Cross-validation
k-fold cross-validation is a technique to evaluate the performance of a model on the test set. It works by splitting the training and testing set into `k` folds, or splits of the data set. Each fold is trained and tested. There is overlap in the training sets, but not the testing sets. 

Five folds is a common choice, but it can be any number. The more folds, the more accurate the evaluation, but the longer it takes to run. With five folds, 20% of the test set is held back each time; with ten folds, 10% of the test set is held back each time. For a smaller dataset, three folds may work better.

The end result is an array containing the k evaluation scores. The mean of the scores is the final score.

#### Cross-validation with linear regression

```python
from sklearn.model_selection import cross_val_score
```

```python
# Initialize a model
model = LinearRegression()

# Implement a 10-fold CV
scores = cross_val_score(model, X, y, cv=10, scoring='neg_mean_squared_error')

# Find the RMSE scores
rmse = np.sqrt(-scores)

# Display the results
print('Reg RMSE: ', np.round(rmse,2))
print('RMSE mean: %0.2f' % (rmse.mean()))
```

#### Cross-validation with XGBoost

```python
# Initialize a model
model = XGBRegressor()

# Implement a 10-fold CV
scores = cross_val_score(model, X, y, cv=10, scoring='neg_mean_squared_error')

# Find the RMSE scores
rmse = np.sqrt(-scores)

# Display the results
print('XGB RMSE: ', np.round(rmse,2))
print('RMSE mean: %0.2f' % (rmse.mean()))
```

### Logistic regression
Logistic regression works similarly to linear regression, but it uses a sigmoid curve that ranges from 0 to 1 rather than summing each term. Values less than 0.5 are matched to 0, and values greater than 0.5 are matched to 1. 

Logistic regression implementation in scikit-learn is nearly the same as linear regression. The main difference is that the predictor column should fit into categories rather than being continuous, and the error should be in terms of accuracy. (This makes explicit scoring parameters unnecessary.)

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Let's build a cross-validation classification function that takes an algorithm as input and returns the accuracy score using cross_val_score().
def cross_val(classifier, num_splits=10):
    model = classifier
    scores = cross_val_score(model, X, y, cv=num_splits)
    print('Accuracy: ', np.round(scores, 2))
    print('Accuracy mean: %0.2f' % (scores.mean()))
```

```python
# Call the CV function with logistic regression
cross_val(LogisticRegression())
```

### XGBoost classifier

```python
from xgboost import XGBClassifier
cross_val(XGBClassifier())
```
